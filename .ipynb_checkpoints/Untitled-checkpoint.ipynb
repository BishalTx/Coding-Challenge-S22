Imporing Libraries

```python
#Importing necessary libraries and API.I am using tensorflow and keras API
import tensorflow as tf
from tensorflow import keras 
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation,Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
```

PrePreocessing

```python
#This command reads csv file using pandas
data = pd.read_csv('mushrooms.csv')
```

```python
#Displaying first five rows of csv file
data.head()
```

```python
#I am checking if there is any null values in the given csv file
data.isnull()
```

```python
mapping = list()
encoder = LabelEncoder()
for column in range(len(data.columns)):
    data[data.columns[column]] = encoder.fit_transform(data[data.columns[column]])
    data_dict = {index:label for index, label in enumerate(encoder.classes_)}
    mapping.append(data_dict)
```

```python
mapping
```

```python
data
```

```python
X_data = data.drop(['class'],axis =1)
Y_data = data['class']
```

```python
scaler = StandardScaler()
X_data = pd.DataFrame(scaler.fit_transform(X_data), columns = X_data.columns)
```

```python
X_data
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, Y_train,Y_test = train_test_split(X_data,Y_data,test_size =0.2,shuffle = True)
```

```python
X_train
```

MOdelBuilding

```python
model = Sequential()
model.add(Dense(32, input_shape= (22,), activation = 'relu'))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(4, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

Training

```python
model.compile(optimizer=Adam(learning_rate=0.0001),loss='binary_crossentropy', metrics=['accuracy'])
trained_model = model.fit(X_train,Y_train, validation_data=(X_test,Y_test),batch_size=10,epochs=30,shuffle=True,verbose=2)
```

```python
test_loss, test_acc = model.evaluate(X_test,Y_test)
```

```python
print(f"Test accuracy {test_acc*100: .2f}%")
```

Visualization

```python
plt.plot(trained_model.history['accuracy'])
plt.plot(trained_model.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc = 'upper right')
```

```python
plt.plot(trained_model.history['loss'])
plt.plot(trained_model.history['val_loss'])
plt.title('Model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc = 'upper right')
```
